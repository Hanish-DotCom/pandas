```python
# 2.
import pandas as pd
df = pd.read_csv('student_placement_data (1).csv')
df.head()
print()
df.shape
print()
df.info()
print()
df.describe()
```

```python
# 3.
print("Missing values:")
print(df.isnull().sum())
print()

print("\nAssignment Completion % Description:")
print(df['Assignment_Completion_%'].describe())
print()

print("\nCommunication Skill (10) Distribution:")
print(df['Communication_Skill_(10)'].value_counts())
print()
```

```python
# 4.

print("Placement Distribution:")
print(df['Placed'].value_counts(normalize=True) * 100)

print("\nAverage Scores by Placement Status:")
print(df.groupby('Placed')[['Assignment_Completion_%', 'Communication_Skill_(10)', 'Content_Score_Avg']].mean().round(2))
```

```python
# 5. correlation between assignment completion % and CTC

print('correlation between assignment completion % and CTC ')
print(df[['Assignment_Completion_%','CTC_LPA']].corr())
print()
# Top 10 performers
print('Top 10 Performers')
print(df.sort_values('Assignment_Completion_%', ascending=False).head(10))
```

```python
# 6. Placement analysis by course
print('Placement Analysis by Course')
print(df.groupby(['Course', 'Placed']).size().unstack(fill_value=0))
print()
# Average CTC per course
print('Average CTC per Course')
print(df.groupby('Course')['CTC_LPA'].mean())
```

```python
# high performers not placed
print('High Performers Not Placed')
print(df[(df['Content_Score_Avg'] > 80) & (df['Communication_Skill_(10)'] > 8) & (df['Placed'] == 'No')])
```
